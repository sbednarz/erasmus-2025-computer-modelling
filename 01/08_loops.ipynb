```python
import numpy as np
import matplotlib.pyplot as plt

a_list = np.array([1.0, 2.0, 3.0, 4.0]) # a parameter
results = []

for a in a_list:
    c = a**3                 # calculations - inline or more complex
    results.append([a, c])   # collect (a, c)

results = np.array(results)

# plot
plt.plot(results[:, 0], results[:, 1], 'bo-')
plt.xlabel("a")
plt.ylabel("c")
plt.title("The influence of a on c")
plt.show()

    
```
